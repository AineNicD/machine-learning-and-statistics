### Classification (supervised learning) <a name = "Classification"></a>

***
Classification is for identifying which category an object belongs to. Examples of such are labeling data, spam detection, image recognition, classification algorithm. 
We classify data by organising it into relevant categories. 
[(Scikit-learn documentation)](https://scikit-learn.org/0.23/_downloads/scikit-learn-docs.pdf)

A comparison of several scikit-learn classifiers on synthetic datasets is shown below.
The purpose of this example is to demonstrate how different classifiers' decision bounds differ.
This example comes from the scikit-learn library, and it excellently illustrates the vast array of tools available in the scikit-learn library.
Toy datasets are used to demonstrate classification methods.
[(Scikit-learn.org)](https://scikit-learn.org/stable/)
The fact that these tools are already included in the package makes scikit-learn an excellent machine learning tool. 

```python
# Code source: Gaël Varoqua, Andreas Müller
# example from scikit-learn library 
# https://scikit-learn.org/stable/auto_examples/classification/plot_classifier_comparison.html#sphx-glr-auto-examples-classification-plot-classifier-comparison-py
# some comments added by me for this demonstration 

# Colormap object generated from a list of colors.
from matplotlib.colors import ListedColormap

# import module for Split arrays or matrices into random train and test subsets
from sklearn.model_selection import train_test_split

# Standardize features by removing the mean and scaling to unit variance.
from sklearn.preprocessing import StandardScaler

# toy datasets to visualize classification algorithms. 
from sklearn.datasets import make_moons, make_circles, make_classification

# This model optimizes the log-loss function using LBFGS or stochastic gradient descent.
from sklearn.neural_network import MLPClassifier

# Classifier implementing the k-nearest neighbors vote.
from sklearn.neighbors import KNeighborsClassifier

# C-Support Vector Classification.
from sklearn.svm import SVC

# Gaussian process classification (GPC) based on Laplace approximation.
from sklearn.gaussian_process import GaussianProcessClassifier

# Radial-basis function kernel (aka squared-exponential kernel).
from sklearn.gaussian_process.kernels import RBF

# A decision tree classifier.
from sklearn.tree import DecisionTreeClassifier

# An AdaBoost classifier.
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

# Gaussian Naive Bayes (GaussianNB).
from sklearn.naive_bayes import GaussianNB

# Quadratic Discriminant Analysis.
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis


# step size in the mesh
h = 0.02  

# the classifier names
names = [
    "Nearest Neighbors",
    "Linear SVM",
    "RBF SVM",
    "Gaussian Process",
    "Decision Tree",
    "Random Forest",
    "Neural Net",
    "AdaBoost",
    "Naive Bayes",
    "QDA",
]

# classifiers
classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="linear", C=0.025),
    SVC(gamma=2, C=1),
    GaussianProcessClassifier(1.0 * RBF(1.0)),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
    MLPClassifier(alpha=1, max_iter=1000),
    AdaBoostClassifier(),
    GaussianNB(),
    QuadraticDiscriminantAnalysis(),
]

# Generate a random n-class classification problem.
X, y = make_classification(
    n_features=2, n_redundant=0, n_informative=2, random_state=1, n_clusters_per_class=1)

# Container for the Mersenne Twister pseudo-random number generator.
rng = np.random.RandomState(2)

# Draw samples from a uniform distribution.
X += 2 * rng.uniform(size=X.shape)
linearly_separable = (X, y)

# the toy datasets
datasets = [
    make_moons(noise=0.3, random_state=0),
    make_circles(noise=0.2, factor=0.5, random_state=1),
    linearly_separable,
]
# plot figure
figure = plt.figure(figsize=(27, 9))
i = 1
# iterate over datasets
for ds_cnt, ds in enumerate(datasets):
    # preprocess dataset, split into training and test part
    X, y = ds
    X = StandardScaler().fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.4, random_state=42
    )

    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # just plot the dataset first
    cm = plt.cm.RdBu
    cm_bright = ListedColormap(["#FF0000", "#0000FF"])
    ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
    if ds_cnt == 0:
        ax.set_title("Input data")
    # Plot the training points
    ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, edgecolors="k")
    # Plot the testing points
    ax.scatter(
        X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.6, edgecolors="k"
    )
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xticks(())
    ax.set_yticks(())
    i += 1

    # iterate over classifiers
    for name, clf in zip(names, classifiers):
        ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
        clf.fit(X_train, y_train)
        score = clf.score(X_test, y_test)

        # Plot the decision boundary. For that, we will assign a color to each
        # point in the mesh [x_min, x_max]x[y_min, y_max].
        if hasattr(clf, "decision_function"):
            Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
        else:
            Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]

        # Put the result into a color plot
        Z = Z.reshape(xx.shape)
        ax.contourf(xx, yy, Z, cmap=cm, alpha=0.8)

        # Plot the training points
        ax.scatter(
            X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, edgecolors="k"
        )
        # Plot the testing points
        ax.scatter(
            X_test[:, 0],
            X_test[:, 1],
            c=y_test,
            cmap=cm_bright,
            edgecolors="k",
            alpha=0.6,
        )

        ax.set_xlim(xx.min(), xx.max())
        ax.set_ylim(yy.min(), yy.max())
        ax.set_xticks(())
        ax.set_yticks(())
        if ds_cnt == 0:
            ax.set_title(name)
        ax.text(
            xx.max() - 0.3,
            yy.min() + 0.3,
            ("%.2f" % score).lstrip("0"),
            size=15,
            horizontalalignment="right",
        )
        i += 1

plt.tight_layout()
plt.show()
```

The plots show training points in solid colors and testing points semi-transparent. The lower right shows the classification accuracy on the test set. [(scikit-learn classifier comparison)](https://scikit-learn.org/stable/auto_examples/classification/plot_classifier_comparison.html#sphx-glr-auto-examples-classification-plot-classifier-comparison-py)

### Linear Regression <a name = "linear-reg"> </a>
***
To compute a simple linear regression model, we can import the linear regression class:

```python
from sklearn.linear_model import LinearRegression
```

Here's an example of regression using sklearn's built-in diabeties dataset from sklearn datasets; it uses the dataset's first feature to depict the data points within the two-dimensional plot.
The straight line illustrates how linear regression seeks to build a straight line as best it can and minimizes residual data points.
Also calculated are the coefficients, residual sum of squares, and coefficient of determination. 
[(Scikit-learn documentation)](https://scikit-learn.org/0.23/_downloads/scikit-learn-docs.pdf)

[https://scikit-learn.org](https://scikit-learn.org/stable/auto_examples/linear_model/plot_ols.html)

```python
# Code source: Jaques Grobler
# https://scikit-learn.org/stable/auto_examples/linear_model/plot_ols.html

#import datasets and linear model
from sklearn import datasets, linear_model

# import metrics 
from sklearn.metrics import mean_squared_error, r2_score

# Load the diabetes dataset
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

# Use only one feature
diabetes_X = diabetes_X[:, np.newaxis, 2]

# Split the data into training/testing sets
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

# Split the targets into training/testing sets
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(diabetes_X_train, diabetes_y_train)

# Make predictions using the testing set
diabetes_y_pred = regr.predict(diabetes_X_test)

# The coefficients
print("Coefficients: \n", regr.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(diabetes_y_test, diabetes_y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(diabetes_y_test, diabetes_y_pred))

# Plot outputs
plt.scatter(diabetes_X_test, diabetes_y_test, color="black")
plt.plot(diabetes_X_test, diabetes_y_pred, color="blue", linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
# generating random data for simple linear regression sklearn example
rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x - 1 + rng.randn(50)
plt.scatter(x, y);
```

Once we have chosen our class of model, we choose model hyperparemeters. 
~~~ 
"An important point is that a class of model is not the same as an instance of a model."
~~~ 
[(jakevdp)](https://github.com/jakevdp/PythonDataScienceHandbook/blob/master/notebooks/05.02-Introducing-Scikit-Learn.ipynb)

For our linear regression example, we can instantiate the LinearRegression class and specify that we would like to fit the intercept using the `fit_intercept` hyperparameter:

```python
# use linear regression class as a model & specify fit intercept
model = LinearRegression(fit_intercept=True)
model
```

When the model is instantiated, the only action is the storing of these hyperparameter values. The Scikit-Learn API makes very clear the distinction between choice of model and application of model to data. We need to massage the data x to make it a matrix of size [n_samples, n_features]. In this case, this amounts to a simple reshaping of the one-dimensional array: [(Scikit-learn documentation)](https://scikit-learn.org/0.23/_downloads/scikit-learn-docs.pdf)

```python
# reshape array 
X = x[:, np.newaxis]
X.shape
```

<b>np.newaxis</b> is used to increase the dimension of the existing array. It uses the slicing operator to recreate the array. The dimension is temporarily added at the position of np.newaxis in the array. ‘None’ can also be used in place of np.newaxis. [(numpy documentation)](https://numpy.org/doc/stable/reference/arrays.indexing.html)

```python
# apply model to data, using fit() method. 
model.fit(X, y)
```

This <b>fit()</b> command performs a series of model-dependent internal computations, the results of which are stored in model-specific attributes that the user can investigate.
All model parameters learned during the <b>fit()</b> process in Scikit-Learn have trailing underscores by convention; for example, in this linear model, we have the following:  [(Scikit-learn documentation)](https://scikit-learn.org/0.23/_downloads/scikit-learn-docs.pdf)

```python
# getting the slope
print( "Slope:",model.coef_)
```

```python
# geting the intercept
print( "intercept:",model.intercept_)
```

The slope and intercept of a basic linear fit to the data are represented by these two parameters.
We can observe that they are extremely near to the input slope of 2 and intercept of -1 when compared to the data description. [(Scikit-learn documentation)](https://scikit-learn.org/0.23/_downloads/scikit-learn-docs.pdf)

The uncertainty in such internal model parameters is a question that commonly arises.
In general, Scikit-Learn doesn't provide tools for drawing conclusions from internal model parameters: interpreting model parameters is more of a statistical modeling problem than a machine learning problem.
Machine learning, on the other hand, is more concerned with what the model predicts. [(Scikit-learn documentation)](https://scikit-learn.org/0.23/_downloads/scikit-learn-docs.pdf)

The objective of supervised machine learning once the model has been trained is to evaluate it based on what it says about new data that was not part of the training set.
The <b>predict()</b> method in Scikit-Learn can be used to accomplish this.
For the sake of this example, we'll use a grid of x values as our "new data," and we'll ask what y values the model predicts: [(Scikit-learn documentation)](https://scikit-learn.org/0.23/_downloads/scikit-learn-docs.pdf)

```python
# Return evenly spaced numbers over specified interval.
xfit = np.linspace(-1, 11)
```

We must again convert these x values into a [n samples, n features] features matrix before feeding it to the model: 

```python
# increase the dimension of above array 
Xfit = xfit[:, np.newaxis]
# predict based on new fitted model 
yfit = model.predict(Xfit)
```

WE can illustrate the findings by plotting the raw data first, followed by this model fit: 

```python
# plotting both raw data and the model fitted
plt.scatter(x, y)
plt.plot(xfit, yfit);
```

Typically, the model's efficacy is assessed by comparing its outcomes to a known baseline.

### Clustering <a name = "Clustering"></a>

***
Clustering algorithms seek to learn, from the properties of the data, an optimal division or discrete labeling of groups of points.  
Clustering of unlabeled data can be performed with the module  [sklearn.cluster.](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.cluster)


[(towardsdatascience.com)](https://towardsdatascience.com/k-means-clustering-with-scikit-learn-6b47a369a83c)

Here is an example of clustering using sklearn's [make_blobs](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html) toy dataset designed for clustering. This example was inspired by Lorraine Li "K-Means Clustering with scikit-learn" post on [(towardsdatascience.com)](https://towardsdatascience.com/k-means-clustering-with-scikit-learn-6b47a369a83c)​.

~~~
sklearn.datasets.make_blobs(n_samples=100, n_features=2, *, centers=None, cluster_std=1.0, center_box=(- 10.0, 10.0), shuffle=True, random_state=None, return_centers=False)
~~~

```python
# import make_blobs toy data set
from sklearn.datasets import make_blobs

# create dataset
X, y = make_blobs(
   n_samples=150, n_features=2,
   centers=3, cluster_std=0.5,
   shuffle=True, random_state=0
)

# plot
plt.scatter(
   X[:, 0], X[:, 1],
   c='white', marker='o',
   edgecolor='black', s=50
)
plt.show()
```

A two-dimensional scatterplot is used to illustrate the dataset, which consists of 150 randomly generated points that are loosely clustered into three zones with increased density. 
We don't have any ground truth category information (knowledge provided as empirical evidence rather than inference) about those samples in real-world clustering applications; otherwise, it would be classified as supervised learning.
As a result, our goal is to group the samples based on feature similarity, which can be accomplished using the k-means algorithm, which can be stated in four steps: [(towardsdatascience)](https://towardsdatascience.com/k-means-clustering-with-scikit-learn-6b47a369a83c)
1. Pick k centroids at random from the sample locations to serve as the first cluster centers.
2. Each sample should be assigned to the nearest centroid (j), j 1,..., k).
3. Move the centroids in the middle of the samples assigned to it.
4. Steps 2 and 3 should be repeated until the cluster assignments remain unchanged or a user-defined tolerance or maximum number of iterations is reached.  [(towardsdatascience.com)](https://towardsdatascience.com/k-means-clustering-with-scikit-learn-6b47a369a83c)
The next challenge is: how can we determine how similar objects are?
The squared Euclidean distance between two points x and y in m-dimensional space is a commonly used distance for grouping samples with continuous features, and it may be defined as the inverse of distance:

![image](images/euclidean_distance.png)

[(image source)](https://towardsdatascience.com/k-means-clustering-with-scikit-learn-6b47a369a83c)


The index j in the previous equation corresponds to the sample points x and y's jth dimension (feature column).
The sample index and cluster index will be denoted by the superscripts I and j, respectively. 
We may characterize the k-means algorithm as a simple optimization problem, an iterative strategy for reducing the within-cluster Sum of Squared Errors(SSE), which is also known as cluster inertia, based on this Euclidean distance metric: 

[(towardsdatascience)](https://towardsdatascience.com/k-means-clustering-with-scikit-learn-6b47a369a83c)

![image](images/cluster_inertia.png)

[(image source)](https://towardsdatascience.com/k-means-clustering-with-scikit-learn-6b47a369a83c)

[(towardsdatascience)](https://towardsdatascience.com/k-means-clustering-with-scikit-learn-6b47a369a83c)
When employing a Euclidean distance metric to apply k-means to real-world data, make sure that the features are measured on the same scale and use z-score standardization or min-max scaling if necessary. 

K-means clustering using scikit-learn

~~~
class sklearn.cluster.KMeans(n_clusters=8, *, init='k-means++', n_init=10, max_iter=300, tol=0.0001, verbose=0, random_state=None, copy_x=True, algorithm='auto')
~~~ 

[(sklearn documentation)](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)
let’s apply k-means clustering to the sample dataset using the KMeans class from scikit-learn's cluster module:

```python
#import k means class
from sklearn.cluster import KMeans

# applying clustering to sample dataset 
km = KMeans(
    n_clusters=3, init='random',
    n_init=10, max_iter=300, 
    tol=1e-04, random_state=0
)
y_km = km.fit_predict(X)
y_km
```

We set the number of required clusters to 3 using the preceding code.
We used n `init=10` to run the k-means clustering algorithms 10 times with different random centroids before selecting the final model with the lowest SSE. We provide the maximum number of iterations for each run using the `max_iter` parameter(here, `300`). 
It's worth noting that scikit-learn's k-means implementation stops short of the maximum number of iterations if it converges.
However, if we choose very large numbers for `max iter`, it is feasible that k-means will not reach convergence for a given run, which can be troublesome (computationally expensive).  [(towardsdatascience)](https://towardsdatascience.com/k-means-clustering-with-scikit-learn-6b47a369a83c)

Choosing bigger values for `tol`, a parameter that regulates the tolerance for changes in the within-cluster sum-squared-error to announce convergence, is one technique to cope with convergence problems.
We chose a tolerance of `1e-04` (= 0.0001) in the preceding code.  [(towardsdatascience)](https://towardsdatascience.com/k-means-clustering-with-scikit-learn-6b47a369a83c)


The fact that one or more clusters can be empty is a concern with k-means.
This issue, however, is addressed in scikit-learn's current k-means implementation.
When a cluster is empty, the method looks for the sample that is farthest away from the empty cluster's centroid.
The centroid will then be reassigned to this distant point.  [(towardsdatascience)](https://towardsdatascience.com/k-means-clustering-with-scikit-learn-6b47a369a83c)


Let's visualize the clusters that k-means detected in the dataset, together with the cluster centroids, now that we've forecast the cluster labels `y km`.
The `cluster centers_` attribute of the fitted `KMeans` object stores these: 

 [(towardsdatascience)](https://towardsdatascience.com/k-means-clustering-with-scikit-learn-6b47a369a83c)









```python
# plot the 3 clusters
plt.scatter(
    X[y_km == 0, 0], X[y_km == 0, 1],
    s=50, c='lightgreen',
    marker='s', edgecolor='black',
    label='cluster 1'
)

plt.scatter(
    X[y_km == 1, 0], X[y_km == 1, 1],
    s=50, c='orange',
    marker='o', edgecolor='black',
    label='cluster 2'
)

plt.scatter(
    X[y_km == 2, 0], X[y_km == 2, 1],
    s=50, c='lightblue',
    marker='v', edgecolor='black',
    label='cluster 3'
)

# plot the centroids
plt.scatter(
    km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='centroids'
)
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

The three centroids were placed at the center of each sphere by k-means in the resulting scatterplot, which appears to be a sensible grouping given this dataset. 

Although k-means performed well on this fake dataset, it's worth noting that one of its drawbacks is that we have to specify the number of clusters, k, before we can figure out what the best k is.
In real-world applications, the number of clusters to choose may not always be evident, especially if we are working with a higher-dimensional dataset that cannot be seen. [(towardsdatascience.com)](https://towardsdatascience.com/k-means-clustering-with-scikit-learn-6b47a369a83c)


For a given task, the elbow technique is a handy graphical tool for estimating the ideal number of clusters k.
We can deduce from this that as k grows, the within-cluster SSE ("distortion") decreases.
This is due to the fact that the samples will be closer to the centroids to which they have been assigned.

[(towardsdatascience.com)](https://towardsdatascience.com/k-means-clustering-with-scikit-learn-6b47a369a83c)

The elbow method's goal is to find the value of k at which the distortion begins to drop most rapidly, which is easier to see if we plot the distortion for various k values: 

[(towardsdatascience.com)](https://towardsdatascience.com/k-means-clustering-with-scikit-learn-6b47a369a83c)

```python
# calculate distortion for a range of number of cluster

# trying out the setting threads in the warning on code cell. 
import os
os.environ["OMP_NUM_THREADS"] = "1" 


distortions = []
for i in range(1, 11):
    km = KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    )
    km.fit(X)
    distortions.append(km.inertia_)

# plot
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

The elbow is placed at k = 3, as can be seen in the resulting plot, indicating that k = 3 is a good option for this dataset.

### Dimensionality reduction <a name ="Dimensionalityreduction"></a>

***
Dimensoinality reduction is used in both supervised and unsupervised learning algorithms. 
It works by deducing the number of random variables to consider by identifying the most significant components of the feature space, keeping them, and discarding the rest.
Principal Component Analysis (PCA), Linear Discriminant Analysis (LDA) and Singular Value Decomposition (SVD) are some of the most common techniques for dimensionality reduction in unsupervised learning.
    
   ##### Principal Component Analysis (PCA) (unsupervised learning ) <a name = "PCA" ></a>
Principal Component Analysis identifies the dataset's "principal" or most influential qualities and generates features from them.  [stackabuse](https://stackabuse.com/dimensionality-reduction-in-python-with-scikit-learn/)
~~~
 sklearn.decomposition.PCA(n_components=None, *, copy=True, whiten=False, svd_solver='auto', tol=0.0, iterated_power='auto', random_state=None)
~~~ 
[sklearn](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html)

 ##### Linear Discriminant Analysis  (LDA)  (supervised learning) <a name = "LDA"></a>
 Fisher's linear discriminant is a method used in statistics and other fields to find a linear combination of features that characterizes or separates two or more classes of objects or events. Linear discriminant analysis (LDA) is a generalization of Fisher's linear discriminant.
The resulting mixture can be used as a linear classifier or, more typically, to reduce dimensionality before further classification.   [analyticsvidhya](https://www.analyticsvidhya.com/blog/2021/08/a-brief-introduction-to-linear-discriminant-analysis/)
~~~
class sklearn.discriminant_analysis.LinearDiscriminantAnalysis(solver='svd', shrinkage=None, priors=None, n_components=None, store_covariance=False, tol=0.0001, covariance_estimator=None
~~~
[sklearn](https://scikit-learn.org/stable/modules/generated/sklearn.discriminant_analysis.LinearDiscriminantAnalysis.html)


The Iris dataset contains four parameters for three types of Iris flowers (Setosa, Versicolor, and Virginica): sepal length, sepal width, petal length, and petal width.

When this data is subjected to Principal Component Analysis (PCA), the combination of attributes (principal components, or directions in the feature space) that account for the most variance in the data is identified.
On the two initial main components, we plot the various samples.

Linear Discriminant Analysis (LDA) seeks to identify the attributes that account for the most variation between classes.

Unlike PCA, LDA is a supervised method that uses a known class label. 
[Scikit-learn documentation](https://scikit-learn.org/0.23/_downloads/scikit-learn-docs.pdf)

```python
# https://scikit-learn.org/stable/auto_examples/decomposition/plot_pca_vs_lda.html#sphx-glr-auto-examples-decomposition-plot-pca-vs-lda-py
# Principal component analysis (PCA).
from sklearn.decomposition import PCA
#Linear Discriminant Analysis.
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# load the iris dataset using sklearn
iris = datasets.load_iris()

# x is the data
X = iris.data
# y is the target for prediction
y = iris.target
target_names = iris.target_names


#fit and transform the data to a PCA class
pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)

#fit and transform the data to a LDA class
lda = LinearDiscriminantAnalysis(n_components=2)
X_r2 = lda.fit(X, y).transform(X)

# Percentage of variance explained for each components
print(
    "explained variance ratio (first two components): %s"
    % str(pca.explained_variance_ratio_)
)

plt.figure()
colors = ["navy", "turquoise", "darkorange"]
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(
        X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=0.8, lw=lw, label=target_name
    )
plt.legend(loc="best", shadow=False, scatterpoints=1)
plt.title("PCA of IRIS dataset")

plt.figure()
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(
        X_r2[y == i, 0], X_r2[y == i, 1], alpha=0.8, color=color, label=target_name
    )
plt.legend(loc="best", shadow=False, scatterpoints=1)
plt.title("LDA of IRIS dataset")

plt.show()
```

<b> Hashing feature transformation using Totally Random Trees </b>

RandomTreesEmbedding allows you to map data to a sparse, high-dimensional representation, which could be useful for classification. The mapping is entirely unsupervised and extremely effective. 
~~~
 sklearn.ensemble.RandomTreesEmbedding(n_estimators=100, *, max_depth=5, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_leaf_nodes=None, min_impurity_decrease=0.0, sparse_output=True, n_jobs=None, random_state=None, verbose=0, warm_start=False)
 ~~~
[(Scikit-learn documentation)](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomTreesEmbedding.html)

The following example illustrates how the transformation can be used for non-linear dimensionality reduction or non-linear classification by visualizing the partitions produced by many trees.

Neighboring points frequently share the same leaf of a tree, and hence share considerable portions of their hashed representation. This allows you to easily separate two concentric circles using shortened SVD and the principal components of the converted data. [Scikit-learn documentation](https://scikit-learn.org/0.23/_downloads/scikit-learn-docs.pdf)

Linear classifiers frequently attain outstanding accuracy in high-dimensional spaces. BernoulliNB is particularly well-suited to sparse binary data. The decision boundary achieved by BernoulliNB in the converted space is compared to an ExtraTreesClassifier forests learned on the original data in the bottom row. [sklearn](https://scikit-learn.org/stable/auto_examples/ensemble/plot_random_forest_embedding.html#sphx-glr-auto-examples-ensemble-plot-random-forest-embedding-py)

```python
# https://scikit-learn.org/stable/auto_examples/ensemble/plot_random_forest_embedding.html#sphx-glr-auto-examples-ensemble-plot-random-forest-embedding-py

# toy dataset from sklearn make_circles
from sklearn.datasets import make_circles
# use RandomTreesEmbedding for transforming data
from sklearn.ensemble import RandomTreesEmbedding, ExtraTreesClassifier
# Dimensionality reduction using truncated SVD (aka LSA).
from sklearn.decomposition import TruncatedSVD
# Naive Bayes classifier for multivariate Bernoulli models.
from sklearn.naive_bayes import BernoulliNB

# make a synthetic dataset
X, y = make_circles(factor=0.5, random_state=0, noise=0.05)

# use RandomTreesEmbedding to transform data
hasher = RandomTreesEmbedding(n_estimators=10, random_state=0, max_depth=3)
X_transformed = hasher.fit_transform(X)

# Visualize result after dimensionality reduction using truncated SVD
svd = TruncatedSVD(n_components=2)
X_reduced = svd.fit_transform(X_transformed)

# Learn a Naive Bayes classifier on the transformed data
nb = BernoulliNB()
nb.fit(X_transformed, y)


# Learn an ExtraTreesClassifier for comparison
trees = ExtraTreesClassifier(max_depth=3, n_estimators=10, random_state=0)
trees.fit(X, y)


# scatter plot of original and reduced data
fig = plt.figure(figsize=(9, 8))

ax = plt.subplot(221)
ax.scatter(X[:, 0], X[:, 1], c=y, s=50, edgecolor="k")
ax.set_title("Original Data (2d)")
ax.set_xticks(())
ax.set_yticks(())

ax = plt.subplot(222)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, s=50, edgecolor="k")
ax.set_title(
    "Truncated SVD reduction (2d) of transformed data (%dd)" % X_transformed.shape[1]
)
ax.set_xticks(())
ax.set_yticks(())

# Plot the decision in original space. For that, we will assign a color
# to each point in the mesh [x_min, x_max]x[y_min, y_max].
h = 0.01
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# transform grid using RandomTreesEmbedding
transformed_grid = hasher.transform(np.c_[xx.ravel(), yy.ravel()])
y_grid_pred = nb.predict_proba(transformed_grid)[:, 1]

ax = plt.subplot(223)
ax.set_title("Naive Bayes on Transformed data")
ax.pcolormesh(xx, yy, y_grid_pred.reshape(xx.shape))
ax.scatter(X[:, 0], X[:, 1], c=y, s=50, edgecolor="k")
ax.set_ylim(-1.4, 1.4)
ax.set_xlim(-1.4, 1.4)
ax.set_xticks(())
ax.set_yticks(())

# transform grid using ExtraTreesClassifier
y_grid_pred = trees.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]

ax = plt.subplot(224)
ax.set_title("ExtraTrees predictions")
ax.pcolormesh(xx, yy, y_grid_pred.reshape(xx.shape),shading='auto')
ax.scatter(X[:, 0], X[:, 1], c=y, s=50, edgecolor="k")
ax.set_ylim(-1.4, 1.4)
ax.set_xlim(-1.4, 1.4)
ax.set_xticks(())
ax.set_yticks(())

plt.tight_layout()
plt.show()
```

k- fold Cross validation 
The following example is a 5 fold cross-validation from [ask python.](https://www.askpython.com/python/examples/k-fold-cross-validation)

```python
#Importing required libraries
# load cancer dataset
from sklearn.datasets import load_breast_cancer
# import kfold class
from sklearn.model_selection import KFold 
# import logistic regresion class
from sklearn.linear_model import LogisticRegression
# import accuracy score 
from sklearn.metrics import accuracy_score
 
#Loading the dataset
data = load_breast_cancer(as_frame = True)
df = data.frame
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
 
#Implementing cross validation
 
k = 5
kf = KFold(n_splits=k, random_state=None)
model = LogisticRegression(solver= 'liblinear')
 
acc_score = []
 
for train_index , test_index in kf.split(X):
    X_train , X_test = X.iloc[train_index,:],X.iloc[test_index,:]
    y_train , y_test = y[train_index] , y[test_index]
     
    model.fit(X_train,y_train)
    pred_values = model.predict(X_test)
     
    acc = accuracy_score(pred_values , y_test)
    acc_score.append(acc)
     
avg_acc_score = sum(acc_score)/k
 
print('accuracy of each fold - {}'.format(acc_score))
print('Avg accuracy : {}'.format(avg_acc_score))
```

Logistic regression was used as the model for binary classification and 5-fold cross-validation was used to cross-validate it. The model's average accuracy was roughly 95.25 percent.

Support Vector Machines are among the most popular supervised machine learning algorithms. They can be used for Classification, Regression and outliers detection.
It can easily handle multiple continuous and categorical variables. SVM constructs a hyperplane in multidimensional space to separate different classes. The core idea of SVM is to find a maximum marginal hyperplane(MMH) that best divides the dataset into classes. ![image](images/SV.png)

Support vectors are the data points that are closest to the hyperplane. By calculating margins, these points will define the seperating line better. The hyperplane is a decision plane which seperates a set of objects having different class. 
The margin is the gap between the two lines on the closest class points. If the margin is larger in between the classes, then it is considered a good margin, a smaller margin is a bad margin. [data camp](https://www.datacamp.com/community/tutorials/svm-classification-scikit-learn-python)

## 2. k-means<a name = "kmeans"></a>
***


![image](images/sklearn_cluster_KMeans.png)

[image source](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html#sklearn.cluster.KMeans)

The Kmeans algorithm is an iterative algorithm that attempts to split a dataset into K separate non-overlapping subgroups (clusters), each of which contains data point belonging to one group.
It attempts to make intra-cluster data points as comparable as possible while maintaining clusters as distinct (far) as possible.
It distributes data points to clusters in such a way that the sum of the squared distances between them and the cluster's centroid which is the arithmetic mean of all the data points in that cluster, is at her minimum.
Within clusters, the less variance there is, the more similar the data points are. 
[(geeksforgeeks)](https://www.geeksforgeeks.org/k-means-clustering-introduction/)
K-Means finds the best centroids by alternating between
1. assigning data points to clusters based on the current centroids 
2. chosing centroids (points which are the center of a cluster) based on the current assignment of data points to clusters. 
[(stanford.edu)](https://stanford.edu/~cpiech/cs221/handouts/kmeans.html)

```python
# load dataset
data = pd.read_csv('./datasets/clustering.csv')
# get the head of the dataset
data.head()
```

For this k-means demonstration, I will be using two variables from the data set, “LoanAmount” and “ApplicantIncome”. 

```python
X = data[["LoanAmount","ApplicantIncome"]]
#Visualise data points
plt.scatter(X["ApplicantIncome"],X["LoanAmount"],c='black')
plt.xlabel('AnnualIncome')
plt.ylabel('Loan Amount (In Thousands)')
plt.show()
```

The first and second steps of the K-Means algorithm involves determining the number of clusters (k) and selecting random centroids for each cluster.
We'll choose three clusters, and then the centroids will be chosen at random from the data: 

```python
# Step 1 and 2 - Choose the number of clusters (k) and select random centroid for each cluster

# number of clusters
K=3

# Select random observation as centroids
Centroids = (X.sample(n=K))
plt.scatter(X["ApplicantIncome"],X["LoanAmount"],c='black')
plt.scatter(Centroids["ApplicantIncome"],Centroids["LoanAmount"],c='red')
plt.xlabel('AnnualIncome')
plt.ylabel('Loan Amount (In Thousands)')
plt.show()
```

The red dots show each cluster's three centroids.
These sites were chosen at random. Different centroids may appear each time the code is run. 
The K-Means Clustering technique will then be implemented after some conditions are established. 

```python
# Step 3 - Assign all the points to the closest cluster centroid
# Step 4 - Recompute centroids of newly formed clusters
# Step 5 - Repeat step 3 and 4

diff = 1
j=0

while(diff!=0):
    XD=X
    i=1
    for index1,row_c in Centroids.iterrows():
        ED=[]
        for index2,row_d in XD.iterrows():
            d1=(row_c["ApplicantIncome"]-row_d["ApplicantIncome"])**2
            d2=(row_c["LoanAmount"]-row_d["LoanAmount"])**2
            d=np.sqrt(d1+d2)
            ED.append(d)
        X[i]=ED
        i=i+1

    C=[]
    for index,row in X.iterrows():
        min_dist=row[1]
        pos=1
        for i in range(K):
            if row[i+1] < min_dist:
                min_dist = row[i+1]
                pos=i+1
        C.append(pos)
    X["Cluster"]=C
    Centroids_new = X.groupby(["Cluster"]).mean()[["LoanAmount","ApplicantIncome"]]
    if j == 0:
        diff=1
        j=j+1
    else:
        diff = (Centroids_new['LoanAmount'] - Centroids['LoanAmount']).sum() + (Centroids_new['ApplicantIncome'] - Centroids['ApplicantIncome']).sum()
        print(diff.sum())
    Centroids = X.groupby(["Cluster"]).mean()[["LoanAmount","ApplicantIncome"]]
```

These numbers may change each time we run the program.
When the centroids do not change after two iterations, the training is terminated.
We started with a value of 1 for the diff, and we calculate it inside the while loop as the difference between the centroids from the previous iteration and the current iteration.
We'll stop training when this disparity reaches zero.
Let's have a look at the clusters we have: 

```python
color=['blue','green','cyan']
for k in range(K):
    data=X[X["Cluster"]==k+1]
    plt.scatter(data["ApplicantIncome"],data["LoanAmount"],c=color[k])
plt.scatter(Centroids["ApplicantIncome"],Centroids["LoanAmount"],c='red')
plt.xlabel('Income')
plt.ylabel('Loan Amount (In Thousands)')
plt.show()
```

Three clusters can be seen plainly here.
The centroid of each cluster is represented by the red dots.

Naive Bayes models are a group of extremely fast and simple classification algorithms that are often suitable for very high-dimensional datasets. They are fast and have so few tuneable parameters which make them very useful for a quick baseline for a classification problem.
```python
from sklearn.datasets import make_blobs
X, y = make_blobs(100, 2, centers=2, random_state=2, cluster_std=1.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu');
```

A fast way to create a simple model is by making the assumption that the data is described by a Gaussian distribution with no covariance between dimensions. We can fit the model by ginding the mean and standard deviation of the points within each label.   [naive-bayes](https://jakevdp.github.io/PythonDataScienceHandbook/05.05-naive-bayes.html)

```python
# https://jakevdp.github.io/PythonDataScienceHandbook/06.00-figure-code.html#Gaussian-Naive-Bayes
# Jacob Vanderplas 

from sklearn.datasets import make_blobs
X, y = make_blobs(100, 2, centers=2, random_state=2, cluster_std=1.5)

fig, ax = plt.subplots()

ax.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
ax.set_title('Naive Bayes Model', size=14)

xlim = (-8, 8)
ylim = (-15, 5)

xg = np.linspace(xlim[0], xlim[1], 60)
yg = np.linspace(ylim[0], ylim[1], 40)
xx, yy = np.meshgrid(xg, yg)
Xgrid = np.vstack([xx.ravel(), yy.ravel()]).T

for label, color in enumerate(['red', 'blue']):
    mask = (y == label)
    mu, std = X[mask].mean(0), X[mask].std(0)
    P = np.exp(-0.5 * (Xgrid - mu) ** 2 / std ** 2).prod(1)
    Pm = np.ma.masked_array(P, P < 0.03)
    ax.pcolorfast(xg, yg, Pm.reshape(xx.shape), alpha=0.5,
                  cmap=color.title() + 's')
    ax.contour(xx, yy, P.reshape(xx.shape),
               levels=[0.01, 0.1, 0.5, 0.9],
               colors=color, alpha=0.2)
    
ax.set(xlim=xlim, ylim=ylim)


```

The eclipses represent the Gaussian model for each label. The larger probability is at the center of the eclipses.  This is a simple receipe to compute the likelihood <b>  P(features | L1) </b> for any data point. This can help us compute quickly the posterior ratio [(naive-bayes)](https://jakevdp.github.io/PythonDataScienceHandbook/05.05-naive-bayes.html) and determine which label is more probably for a given point. 

This procedure is implemented in Scikit-Learn's sklearn.naive_bayes.GaussianNB estimator:

![image](images/GaussianNB.png)

[image source](https://scikit-learn.org/0.15/modules/generated/sklearn.naive_bayes.GaussianNB.html)

```python
# https://jakevdp.github.io/PythonDataScienceHandbook/06.00-figure-code.html#Gaussian-Naive-Bayes
# Jacob Vanderplas 
# some tweeking by me for demonstration but still mainly code from source


from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X, y);

# generate data and predict the label
rng = np.random.RandomState(0)
Xnew = [-6, -14] + [14, 18] * rng.rand(2000, 2)
ynew = model.predict(Xnew)

# plot data to get idea of where the decision boundry is 
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
lim = plt.axis()
plt.scatter(Xnew[:, 0], Xnew[:, 1], c=ynew, s=20, cmap='RdBu', alpha=0.1)
plt.axis(lim);

# This Bayesian approach has the added benefit of allowing for probabilistic classification, 
# which we can compute with the predict proba method:
yprob = model.predict_proba(Xnew)
print ("Predict proba method:", yprob[-8:].round(2))
```

The posterior probability of the first and second labels are given in the columns. Bayesian techniques like this can be beneficial if you're seeking for estimates of uncertainty in your classification.

The final classification will only be as good as the model assumptions that lead up to it, which is why Gaussian naive Bayes frequently fails to deliver satisfactory results. Even so, in many cases—particularly as the number of features grows—this assumption isn't enough to make Gaussian naive Bayes an ineffective strategy.
